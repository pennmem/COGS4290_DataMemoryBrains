## Inter-Response Time Curves
An interresponse time (IRT) is the amount of time that elapses between two consecutive responses. If responses occur at times t₁, t₂, t₃, …, then the kth interresponse time is simply tₖ₊₁ − tₖ. In memory experiments such as free recall, this corresponds to the pause between recalling one item and recalling the next. An interresponse time curve summarizes how these pauses change as a sequence of responses unfolds. 

**See in FHM Figure 6.11**

The defining empirical feature of an IRT curve in many cognitive tasks, especially free recall, is that IRTs increase with output position: early responses occur quickly, and later responses are separated by longer and longer delays. Early in a recall sequence, highly accessible memories are readily available and closely matched to the current retrieval context, leading to short interresponse times. As these items are exhausted, the retrieval system must search more broadly, rely on weaker contextual cues, or recover from failed retrieval attempts

## Lag-CRP
The Lag Conditional Response Probability (**Lag-CRP**) is a calculation measuring the temporal organization of memory.  When we transition freely from one recall to another, are we more likely to transition between items that were presented close together in time or far apart in time?  The Lag-CRP calculation is essentially considering each **lag**, or the serial position jump between adjacent recalls, and then dividing the number of actual transitions bt the number of possible transitions, giving a **conditional response probability**. 

In free recall experiments, items are presented in a sequence during study, and recall can occur in any order. The lag between two recalled items is defined as the difference in their original serial positions at study. If item j was studied immediately after item i, the lag is +1; if it was studied immediately before item i, the lag is −1. Larger absolute lag values correspond to items that were farther apart in time during encoding.

A Lag CRP curve plots the probability of making a recall transition as a function of this lag. On the x-axis is lag (negative to positive), and on the y-axis is the conditional probability that the next recalled item comes from that lag, given the set of items that are still available to be recalled at that moment. The “conditional” part is crucial: the probability is computed relative to which lags are actually possible at each transition, so the curve reflects preference rather than simple frequency.

## PLI Recency and the PLI Recency Curve 
A prior-list intrusion (PLI) recency curve describes how often intrusions during recall come from lists studied further back in time, and how that likelihood changes as a function of list lag. In multi-list free recall paradigms, participants study and recall a sequence of lists. A PLI occurs when, during recall of the current list, a participant recalls an item that belonged to a previous list instead of the target list.

To construct a PLI recency curve, each intrusion is assigned a list lag based on how many lists back it originated from. An item recalled from the immediately preceding list has a lag of −1, from two lists back has a lag of −2, and so on. The curve then plots the probability (or frequency, often normalized) of PLIs as a function of this list lag.

The defining feature of the PLI recency curve is a strong recency gradient. Intrusions are far more likely to come from recent lists than from lists further in the past. Items from the immediately preceding list dominate, and the probability of intrusions drops off rapidly as the number of intervening lists increases. This produces a steeply decreasing curve as list lag becomes more negative.
